```python
vals = state_demographic.head(10)[['LALVOTERID',
                      'Residence_Addresses_Latitude',
                      'Residence_Addresses_Longitude']]
vals
```

```python
vals['Residence_Addresses_Latitude'] = pd.to_numeric(vals['Residence_Addresses_Latitude']) 
vals['Residence_Addresses_Longitude'] = pd.to_numeric(vals['Residence_Addresses_Longitude']) 
vals.dtypes
```

```python
jittered_lat = vals.jitter(
    column_name='Residence_Addresses_Latitude',
    dest_column_name='jittered_lat',
    scale=0.2
)
jittered_lat
```

```python
jittered_long = vals.jitter(
    column_name='Residence_Addresses_Longitude',
    dest_column_name='jittered_long',
    scale=0.2
)
jittered_long
```

```python
def handle_location_jittering(dataset):
    dataset['Residence_Addresses_Latitude'] = pd.to_numeric(dataset['Residence_Addresses_Latitude']) 
    dataset['Residence_Addresses_Longitude'] = pd.to_numeric(dataset['Residence_Addresses_Longitude']) 
    dataset.jitter(
    column_name='Residence_Addresses_Latitude',
    dest_column_name='jittered_lat',
    scale=0.2)
    
    dataset.jitter(
    column_name='Residence_Addresses_Longitude',
    dest_column_name='jittered_long',
    scale=0.2)
    
    dataset = dataset.drop(columns=['Residence_Addresses_Latitude', 'Residence_Addresses_Longitude'])
    
    return dataset
    
jittered_data = handle_location_jittering(state_demographic)
jittered_data
```
